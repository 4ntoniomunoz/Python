# Estructuras de Datos
Las estructuras de datos son formas de organizar informacion para trabajar con ellas de manera eficiente.

- Listas -> Ordenadas , modificadas y permiten duplicados
- Tuplas -> Ordenadas , inmutables y permiten duplicados
- sets -> No ordenados y no se permiten duplicados
- Diccionarios -> Pares clave:valor

Nota:
- Pilas (FILO -> First in, Last out)
- Colas (FIFO -> First in, First out
- Listas enlazadas (o doblemente enlazadas)

```python
#Ejemplo
import random

opciones = ["piedra", "papel", "tijeras"]

# Ciclo principal del juego
while True:
    # Ciclo para validar la entrada del usuario
    while True:
        eleccion = input("\nElige piedra, papel o tijeras (o escribe 'salir' para terminar): ")
        eleccion = eleccion.lower() # Convertir a min√∫sculas para manejar 'SALIR' o 'Piedra'

        if eleccion in opciones or eleccion == "salir":
            break
        else:
            print("Opci√≥n no v√°lida, intenta de nuevo.")

    # Si el usuario eligi√≥ salir, rompemos el ciclo principal
    if eleccion == "salir":
        print("¬°Gracias por jugar!")
        break

    aleatoria = random.choice(opciones)
    print("La computadora eligi√≥:", aleatoria)

    # Determinamos el ganador
    if eleccion == aleatoria:
        print("Empate")
    elif (eleccion == "piedra" and aleatoria == "tijeras") \
        or (eleccion == "papel" and aleatoria == "piedra") \
        or (eleccion == "tijeras" and aleatoria == "papel"):
        print("¬°Ganaste!:))")
    else:
        print("Perdiste:((")
```

## SETS

Los sets son colecciones SIN orden y que NO permiten elementos duplicados.

```python
nombre = {"Angel","Eliud","Diana","Juan","Juan"}
print(nombre)
print(type(nombre))
```

```python
distintos_datos = {"Angel", 1 ,"Diana","Juan", 4.8}
distintos_datos
```

```python
# Add
nombre.add("Leonardo")
nombre
```

```python
# Union
matutino = {20322,22183,12355,134123}
vespertino = {20322,31231,123123,1233213}
#Union pero con .union()
total_matriculas = matutino.union(vespertino)
print(total_matriculas)
# Union pero con |
total_matriculas = matutino | vespertino
print(total_matriculas)

```

```python
# Interseccion
ca = {1234,1234,5123,2134,1634,87543,213567}
al = {1234,1111,2222,3333,444}
#interseccion con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
#interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)

# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
nombre = {"Angel","Eliud","Diana","Juan","Juan"}
#remove, pop
#nombre.remove("Angel") #Puedes eliminar el que tu quieras
print(nombre)
nombre.pop() #Elimina un elemento de manera aleatoria
print(nombre)
```

## Diccionarios
Los diccionarios son clecciones no ordenadas de pares clave:valor



```python
alumno = {
    #clave     valor
    "nombre": "Eliud Sanchez",
    "edad": 21,
    "carrera": "LCC",
    "semestre": 7,
    "promedio": 88,
    "calificaciones": [100,94, 94 ,97, 100]

}
print(type(alumno))
alumno
```

```python
# üçé 1. INICIALIZACI√ìN DE DATOS (Variables y Estructuras)
# --- LISTA: Nombres de productos ---
PRODUCTOS = ["manzana", "banana", "leche", "pan"]

# --- TUPLA: Precios correspondientes ---
PRECIOS = (10, 5, 20, 15)

# --- SET: Productos en oferta (para b√∫squeda r√°pida) ---
OFERTAS = {"manzana", "pan"}

# --- DICCIONARIO: Inventario (Clave: producto, Valor: {precio, stock}) ---
INVENTARIO = {
    PRODUCTOS[i]: {"precio": PRECIOS[i], "stock": 10}
    for i in range(len(PRODUCTOS))
}


# üõí 2. FUNCI√ìN DE VISUALIZACI√ìN (Muestra el estado actual del inventario)
def mostrar_inventario(inventario_actual, productos_oferta):
    """Imprime el inventario con precios, stock y estado de oferta."""
    print("\n--- INVENTARIO DE TIENDA ---")
    for producto in PRODUCTOS:
        info = inventario_actual.get(producto, None)
        if not info:
            continue

        es_oferta = "OFERTA!!!" if producto in productos_oferta else "Grandes precios!"
        print(
            f"{producto.title():<8} - ${info['precio']:<4} - Stock: {info['stock']:<2} -> {es_oferta}"
        )


# üõçÔ∏è 3. L√ìGICA PRINCIPAL (Mostrar y Comprar)
mostrar_inventario(INVENTARIO, OFERTAS)

# Compra simple
compra = input("\n¬øQu√© producto quieres comprar? ").strip().lower()

if compra in INVENTARIO and INVENTARIO[compra]["stock"] > 0:
    INVENTARIO[compra]["stock"] -= 1
    print(
        f"Compraste 1 {compra}. Stock restante: {INVENTARIO[compra]['stock']}"
    )
    # Opcional: Mostrar el inventario actualizado despu√©s de la compra
    # mostrar_inventario(INVENTARIO, OFERTAS)
else:
    print("Producto no disponible o sin stock.")
```
