## Operaciones Matematicas

```python
# 4to Nivel de importancia
# Suma Ordinaria
print("Suma Ordinaria: ")
print(3 + 4) # un 7
variable1,variable2 = 20,40 #60
print(variable1 + variable2)

# Resta Ordinaria
print("Resta Ordinaria: ")
print(2 - 5) # -3

# Declaracion de variables
print("Operacion Matematico: ")
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) # 4.6
```

```python
# 3er Nivel de importancia
# Multiplicacion Ordinaria
print("Multiplicacion Ordinaria")
M1, M2 = 2,3.14
print(M1 * M2) #6.28
#Division ordinaria
print("------------------")
print("Division Ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
# Division de Piso
print("-----------------")
print("Division de piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)

print( 6 / 2 * 3)
```

```python
# Potencias
print("Potencias")
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2) #49
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
A # A significa print,
  # en este modo o pagina o notebook funciona print como A
```
