# Funciones que llaman a otras Funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adiós, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse (nombre)
  return saludo + "\n" + despedida

print(mensaje_completo("Ana"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabor disponible:", ",", sabores)
  # lower() hace minusculas las palabras
  # strip() elimina los espacios tanto antes como despues
  sabor = input("Elige un sabor: ").strip().lower()
  #si sabor NO esta en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente): ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("¿Cuántas bolas de nieve?: " ))
  while cantidad <= 0:
    cantidad = int(input("¿Cuántas bolas de nieve? (Ingresa un valor valido): "))
  return cantidad
#Funcion Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("¿Cuantos clientes atenderas?: "))
for i in range(total_pedidos):
  pedir_helado()
```

#Funciones Lambda
Una funciona lambda esa una funcion acortada para escribir una funcion de esta forma:

nombreFuncion = lambda parametros: accion

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco mas limitadas

```python
sumar = lambda x,y : x+y
print (sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1 - desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0],",el descuento que se hizo fue de: ", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
def calcular_promedio_y_maximo(calificaciones):
    """
    Calcula el promedio y la calificación máxima de una lista.
    """
    if not calificaciones:
        return 0, 0  # Manejar lista vacía

    # El promedio se calcula sumando todos los elementos y dividiendo por la cantidad
    promedio = sum(calificaciones) / len(calificaciones)

    # Usamos la función max() incorporada para obtener el mayor
    maximo = max(calificaciones)

    return promedio, maximo

# --- PROGRAMA PRINCIPAL ---

calificaciones = []
continuar = "si"

print("--- Calculadora de Calificaciones ---")

while continuar == "si":
    # Bloque de validación para la entrada de calificación
    while True:
        try:
            # Pedir la calificación y asegurarse de que es un número entero
            calif = int(input("Indique la calificación (0-100): "))

            # Validar que esté en el rango permitido
            if 0 <= calif <= 100:
                calificaciones.append(calif)
                break  # Salir del bucle de entrada de calificación
            else:
                print("Error: Ingrese un valor entre 0 y 100.")
        except ValueError:
            print("Error: Ingrese un valor numérico entero.")

    # Preguntar si desea continuar, solo si ya hay calificaciones ingresadas
    if len(calificaciones) > 0:
        continuar = ""
        while continuar not in ("si", "no"):
            continuar = input("\n¿Desea ingresar otra calificación? (si/no): ").lower()
    else:
        # En caso de que el usuario haya fallado la primera entrada
        continuar = "si"

# --- MOSTRAR RESULTADOS ---

if calificaciones:
    promedio, mayor = calcular_promedio_y_maximo(calificaciones)

    print("\n--- Resultados ---")
    print(f"Número de calificaciones ingresadas: {len(calificaciones)}")
    print(f"Promedio: {promedio:.2f}")  # Mostrar promedio con 2 decimales
    print(f"Calificación mayor: {mayor}")
else:
    print("\nNo se ingresaron calificaciones.")
```
